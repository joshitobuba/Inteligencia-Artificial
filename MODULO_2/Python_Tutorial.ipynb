# 12.- CICLOS FOR

```python
# Los ciclos for en Python son estructuras de control que se utilizan para iterar
# sobre una secuencia de elementos. Esto puede incluir listas, tuplas,
# diccionarios, sets, strings o cualquier objeto iterable. El bucle for permite
# ejecutar un bloque de código repetidamente, una vez por cada elemento en la
# secuencia.

# La sintaxis básica de un bucle for en Python es:

for elemento in secuencia:
    # Hacer algo con 'elemento'

# Donde elemento es la variable que toma el valor del elemento actual en la
# iteración, y secuencia es la colección de elementos a iterar.

# Por ejemplo, si tienes una lista de frutas, puedes usar un bucle for para
# imprimir el nombre de cada fruta:

frutas = ["manzana", "banana", "cereza"]
for fruta in frutas:
    print(fruta)

# Este bucle se repetirá tres veces, ya que hay tres elementos en la lista frutas.
# En cada iteración, la variable fruta tomará el valor de uno de los elementos de
# la lista ("manzana", luego "banana", y finalmente "cereza"), y ejecutará el
# bloque de código dentro del bucle, que en este caso es simplemente imprimir
# el nombre de la fruta.

# Los bucles for son una herramienta fundamental en Python que facilitan la
# realización de tareas repetitivas sobre colecciones de datos.
```

```python
# La comprensión de listas (list comprehension) en Python es una forma concisa y
# conveniente de crear listas. Esencialmente, es una sintaxis que permite
# representar operaciones que involucran listas de una manera que es más compacta
# y a menudo más legible que usar bucles for convencionales.

# La sintaxis básica de una comprensión de lista es:

expresion for elemento in iterable if condicion

# expresión: Es lo que se va a hacer con cada elemento del iterable, por ejemplo,
# podría ser una operación matemática o una llamada a función.
# elemento: Es el nombre de la variable que toma el valor del elemento actual en
# la iteración.
# iterable: Es una colección de elementos que puede ser una lista, un rango, un
# string, etc.
# condición: Es opcional y se usa para incluir en la nueva lista solo aquellos
# elementos que cumplan con la condición.
```

```python
# Dada la lista: [1, 2, 3, 4, 5]
# La comprensión de lista debería producir: [1, 9, 25]

# Lista original de números
numeros = [1, 2, 3, 4, 5]

# Comprensión de lista para obtener los cuadrados de los números impares
cuadrados_impares = [n**2 for n in numeros if n % 2 != 0]

print(cuadrados_impares)

```

```python
# El bucle while es útil cuando no se sabe de antemano cuántas veces se necesita
# ejecutar el bucle, a diferencia de un bucle for, que se utiliza cuando se conoce
#  el número de iteraciones o se itera sobre una colección de elementos.

# El bucle while en Python es una estructura de control de flujo que permite
# repetir la ejecución de un bloque de código mientras una condición específica
# sea verdadera. En cada iteración del bucle, la condición es evaluada; si la
# condición es verdadera, el código dentro del bucle se ejecuta. Si la condición
# es falsa, el bucle deja de ejecutarse y el flujo del programa continúa con
# cualquier código que esté después del bucle.

# Aquí tienes un resumen de su funcionamiento:

# Condición inicial: Antes de entrar al bucle, se establece una condición inicial
# que en principio es verdadera.
# Evaluación de la condición: El bucle while evalúa la condición; si es verdadera,
# se ejecuta el bloque de código.
# Ejecución del bloque de código: Se ejecuta todo el código dentro del bloque del
# bucle while.
# Actualización de la condición: Generalmente, dentro del bucle, algo debe cambiar
# para que eventualmente la condición se vuelva falsa. Si no hay nada que cambie
# la condición, podría resultar en un bucle infinito.
# Repetición o terminación: Después de ejecutar el bloque de código, la condición
# se evalúa nuevamente. Si sigue siendo verdadera, el bucle se repite. Si es
# falsa, el bucle termina.



```

```python
# Escribe un programa en Python que haga lo siguiente:

# Inicia una variable contador en 0.
# Usa un bucle while para incrementar contador en 1 en cada iteración.
# El bucle debe detenerse cuando contador sea igual a 5.
# En cada iteración, imprime el valor actual de contador.
# Este es un ejercicio muy básico para demostrar el uso de un bucle while y cómo
# se puede controlar su flujo con una condición simple.


# Inicia el contador en 0
contador = 0

# Bucle while que se ejecuta mientras el contador sea menor que 5
while contador < 5:
    # Imprime el valor actual del contador
    print("Contador actual:", contador)

    # Incrementa el contador en 1
    contador += 1

# Imprime un mensaje cuando el bucle termina
print("El bucle ha terminado y el contador es:", contador)
```

```python
#crea una clase Rectangulo que tenga atributos de ancho y alto y metodos
#que calculen el area y el perimetro del rectangulo
class Rectangulo:
    def __init__(self, ancho, altura):
        self.ancho = ancho
        self.altura = altura

    def calcular_area(self):
        return self.ancho * self.altura

    def calcular_perimetro(self):
        return 2 * (self.ancho + self.altura)

rectangulo1 = Rectangulo(6, 3)
area = rectangulo1.calcular_area()
perimetro = rectangulo1.calcular_perimetro()

print('Area: ',area)
print('Perimetro: ',perimetro)
```

```python
# Realizar un programa en el cual se declaren dos valores enteros por
# teclado utilizando el método __init__. Calcular después la suma, resta,
#  multiplicación y división. Utilizar un método para cada una
# e imprimir los resultados obtenidos. Llamar a la clase Calculadora.

# creamos la clase
class Calculadora:
	# iniciamos con el método __init__
	def __init__(self):
		self.valor1=int(input("Ingrese el primer valor: "))
		self.valor2=int(input("Ingrese el segundo valor: "))

	# función para sumar
	def sumar(self):
		sumar=self.valor1+self.valor2
		print("El resultado de la suma de los valores es: ",sumar)

	# función para restar
	def restar(self):
		restar=self.valor1-self.valor2
		print("El resultado de la resta de los valores es: ",restar)

	# función para calcular el producto
	def multiplicar(self):
		pro=self.valor1*self.valor2
		print("El resultado de la multiplicación de los valores es: ",pro)

	# función para dividir
	def dividir(self):
		div=self.valor1/self.valor2
		print("El resultado de la división de los valores es: ",div)

	# función para imprimir
	def imprimir(self):
		print("Los valores son: ")
		print("Valor 1: ",self.valor1)
		print("Valor 2: ",self.valor2)


# bloque principal
calcular=Calculadora()
calcular.imprimir()
calcular.sumar()
calcular.restar()
calcular.multiplicar()
calcular.dividir()
```

```python
# Realizar un programa que conste de una clase llamada Alumno que tenga como
# atributos el nombre y la nota del alumno. Definir los métodos para inicializar
# sus atributos, imprimirlos y mostrar un mensaje con el resultado de la nota y
# si ha aprobado o no.´

# inicializamos la clase
class Alumno:
    # inicializamos los atributos
    def inicializar(self,nombre,nota):
        self.nombre=nombre
        self.nota=nota


    # función para imprimir los datos
    def imprimir(self):
        print("Nombre: ",self.nombre)
        print("Nota: ",self.nota)


    # función para obtener el resultado
    def resultado(self):
        if self.nota < 5:
            print("El alumno ha suspendido")
        else:
            print("El alumno ha aprobado")



# bloque principal
# creamos los nuevos objetos
alumno1=Alumno()
alumno2=Alumno()

# inicializamos los atributos
alumno1.inicializar("Ivan",6)
alumno2.inicializar("Juan",3)

# imprimimos los datos y resultados de cada alumno
alumno1.imprimir()
alumno1.resultado()
alumno2.imprimir()
alumno2.resultado()
```

```python
# Una función lambda en Python es una forma breve de declarar funciones anónimas,
# es decir, funciones sin nombre. Son útiles para operaciones sencillas que pueden
# ser escritas en una sola línea.

#La sintaxis de una función lambda es:

lambda argumentos: expresion

# Por que usar funcion lambda?
# Las ventajas de lambda se ven sobre todo cuando es una funcion anonima
# dentro de otra funcion

# Digamos que definimos una funcion que toma un  argumento, y ese argumento
# sera multiplicado por un numero desconocido:
```

```python
#Un ejemplo sencillo

# Función lambda para sumar dos números
suma = lambda x, y: x + y

# Uso de la función lambda
resultado = suma(5, 3)
print(resultado)  # Salida: 8

# En este ejemplo, se define una función lambda que toma dos argumentos x e y y
# devuelve su suma. Luego, se utiliza esta función para sumar 5 y 3, y se imprime
# el resultado, que es 8.
```

```python
#Otro ejemplo
x = lambda a, b : a * b
print(x(5, 6))
```

```python
x = lambda a, b, c : a + b + c
print(x(5, 6, 2))
```

```python
# Función lambda para calcular el cuadrado de un número
cuadrado = lambda x: x * x

# Uso de la función lambda
resultado = cuadrado(4)
print(resultado)  # Salida: 16
```

```python
#En este caso, utilizaremos una función lambda para revertir una cadena de texto:

revertir_cadena = lambda s: s[::-1]

# Uso de la función lambda
resultado = revertir_cadena("Hola")
print(resultado)  # Salida: aloH
```
