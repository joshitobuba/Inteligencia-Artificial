# CREAR GRAFICA BASICA


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Podemos crear graficas tan sencillas como esta
x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

plt.plot(x, y)

plt.show()
```

```python
#o tambien podemos crear graficas tan impresionantes como esta

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
X = np.arange(-5, 5, 0.15)
Y = np.arange(-5, 5, 0.15)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis')
plt.show()
```

```python
# o como esta
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

```python
#GRAFICA DE CIRCULOS O SCATTER

#La x muestra la edad de 13 coches
#La y muestra la velocidad a la que van por una carretera

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]


plt.scatter(x, y)

plt.show()

#nos muestra que cuanto ams nuevo el coche a mas velocidad va
```

```python
#dia 1, la edad y velociadad de 13 coches:
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
plt.scatter(x, y)

#dia 2, la edad y velocidad de 15 coches:
x = [2,2,8,1,15,8,12,9,7,3,11,4,7,14,12]
y = [100,105,84,105,90,99,90,95,94,100,79,112,91,80,85]
plt.scatter(x, y)

plt.show()

#por default los colores son azul y naranja
#el nuevo plot nos indica tambien que cuanto mas nuevo el coche
# a mas velocidad conduce
```

```python
#podemos cambiar el color de los puntos

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
plt.scatter(x, y, color='black')


x = [2,2,8,1,15,8,12,9,7,3,11,4,7,14,12]
y = [100,105,84,105,90,99,90,95,94,100,79,112,91,80,85]
plt.scatter(x, y, color='green')

plt.show()
```

```python
#podemos darle un color a cada punto del grafico

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

```python
#puedes darle un tamao a los puntos

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

```python
#tambien puedes ajustar la intensidad del color con el argumento alpha
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.80)

plt.show()
```

```python
#GRAFICO DE BARRAS

#vamos a ver otro tipo de grafico de barras, se usa el metodo bar()
x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

plt.bar(x, y)

plt.show()
```

```python
#el grafico tambien puede tener las barras horizontales, usamos el metodo
#barh()

x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

plt.barh(x, y)

plt.show()
```

```python
#podemos cambiar el color de las barras

x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

plt.barh(x, y, color='y')

plt.show()
```

```python
#tambien podemos cambiar el ancho de las barras

x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

plt.bar(x, y, width=0.3)

plt.show()
```

```python
#para graficos de barras horizontales en vez de ancho usamos alto 'height'

x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

plt.barh(x, y, height= 0.4)

plt.show()
```

```python
#GRAFICA HISTOGRAMA

#Otro tipo de grafico muy comun es el histograma
#El histograma muestra que muestra el numero de observaciones dentro de
#cada intervalo
#Muestra distribuciones de frecuencia

#Si por ejemplo preguntamos la altura a 25 personas, podriamos obtener
#un histograma como este

x= np.random.normal(170, 10, 250)
#170 es el valor medio y la desviacion estandar es 10, 250 es el numero de
#observaciones

plt.hist(x)
plt.show()
```

```python
#GRAFICA DE TARTA O DE PIZZA

#tambien podemos crear un grafico de pizza
x= [1, 2, 3, 4, 10]
#se mueve de derecha a izquierda y empieza por la mitad de la pizza


plt.pie(x)

plt.show()
```

```python
#vamos a aadirle etiquetas a los trozos de pizza

x= [1, 2, 3, 4, 10]
etiquetas=['Peras', 'Platanos', 'Naranjas', 'Manzanas', 'Uvas']

plt.pie(x, labels= etiquetas)

plt.show()
```

```python
#puede que queramos destacar uno de los trozos de pizza usando el parametro
#explode

x= [1, 2, 3, 4, 10]
etiquetas=['Peras', 'Platanos', 'Naranjas', 'Manzanas', 'Uvas']
destacar= [0.4, 0, 0.4, 0, 0]

plt.pie(x, labels= etiquetas, explode=destacar)

plt.show()
```

```python
#le podemos aadir sombras a la grafica usando el parametro shadow como True
x= [1, 2, 3, 4, 10]
etiquetas=['Peras', 'Platanos', 'Naranjas', 'Manzanas', 'Uvas']
destacar= [0.4, 0, 0.4, 0, 0]

plt.pie(x, labels= etiquetas, explode=destacar, shadow=True)

plt.show()

```

```python
#le podemos cambiar el color a cada trozo de pizza

x= [1, 2, 3, 4, 10]
etiquetas=['Peras', 'Platanos', 'Naranjas', 'Manzanas', 'Uvas']

colores= ['Green', 'Yellow', 'Orange', 'Red', 'Purple']

plt.pie(x, labels= etiquetas, colors=colores)

plt.show()
```

```python
#tambien le podemos aadir un cartelito a la grafica con el metodo
#legend()

x= [1, 2, 3, 4, 10]
etiquetas=['Peras', 'Platanos', 'Naranjas', 'Manzanas', 'Uvas']


plt.pie(x, labels= etiquetas)

plt.legend()
plt.show()
```

```python
#para aadirle un encabezado al cartelito usamos otra vez el metodo
#legend() pero con el parametro title dentro del parentesis

x= [1, 2, 3, 4, 10]
etiquetas=['Peras', 'Platanos', 'Naranjas', 'Manzanas', 'Uvas']


plt.pie(x, labels= etiquetas)
plt.legend(title='FRUTAS')

plt.show()
```

# DISEO DE MARKERS


```python
#los markers son los puntos sobre los que despues se dibuja la linea

x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

#ms es el tamao del marker
plt.plot(x, y, marker='|', ms=10)

plt.show()
```

```python
x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

#mec es el color del borde del marker
plt.plot(x, y, marker='*', ms=10, mec='r')

plt.show()
```

```python
x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

#mfc es el color del interior del marker (black, yellow, red, orange, blue, pink, green, etc)
#algunos colores, no todos aceptan poner solo la primera letra de su nombre
plt.plot(x, y, marker='*', ms=10, mfc='pink')

plt.show()
```

## ETIQUETAS Y TITULOS

```python
#vamos a usar xlabel()    ylabel()    title
```

```python
x=[1, 3, 5, 8, 10, 12, 15, 18]
y=[0.3, 0.5, 0.7, 1.10, 1.30, 1.50, 1.60, 1.75]

plt.plot(x, y)

plt.title('Crecimiento del ser humano')
plt.xlabel('Edad')
plt.ylabel('Altura')

plt.show()
```

```python
#Podemos cambiarle el tipo de letra (font) a los titulos y etiquetas
font_algerian = {'fontname': 'Algerian'}
x=[1, 3, 5, 8, 10, 12, 15, 18]
y=[0.3, 0.5, 0.7, 1.10, 1.30, 1.50, 1.60, 1.75]


font1= {'fontfamily':'fantasy', 'color':'green', 'size':15}
font2= {'fontfamily':'serif','color':'purple', 'size':10}
plt.plot(x, y)

plt.title('Crecimiento del ser humano', fontdict=font1)
plt.xlabel('Edad', fontdict=font2)
plt.ylabel('Altura', fontdict=font2)

plt.show()
```

```python
#podemos poner el titulo del grafico a la izquierda, derecha o por default
#en el centro, usamos el 'loc' y el 'right' y 'left'

x=[1, 3, 5, 8, 10, 12, 15, 18]
y=[0.3, 0.5, 0.7, 1.10, 1.30, 1.50, 1.60, 1.75]


font1= {'family':'serif', 'color':'green', 'size':15}
font2= {'family':'serif', 'color':'purple', 'size':10}
plt.plot(x, y)

plt.title('Crecimiento del ser humano', fontdict=font1, loc='left')
plt.xlabel('Edad', fontdict=font2)
plt.ylabel('Altura', fontdict=font2)

plt.show()
```

```python
#podemos aadirle cuadriculas al grafico con el metodo grid()

x=[1, 3, 5, 8, 10, 12, 15, 18]
y=[0.3, 0.5, 0.7, 1.10, 1.30, 1.50, 1.60, 1.75]



plt.plot(x, y)

plt.title('Crecimiento del ser humano')
plt.xlabel('Edad')
plt.ylabel('Altura')

plt.grid()
plt.show()
```

```python
#el metodo anterior se puede aadir solo a las filas o solo a las columnas
x=[1, 3, 5, 8, 10, 12, 15, 18]
y=[0.3, 0.5, 0.7, 1.10, 1.30, 1.50, 1.60, 1.75]



plt.plot(x, y)

plt.title('Crecimiento del ser humano')
plt.xlabel('Edad')
plt.ylabel('Altura')

plt.grid(axis='x')
#o plt.grid(axis='y')

plt.show()
```

```python
#tambien podemos disear las cuadriculas

x=[1, 3, 5, 8, 10, 12, 15, 18]
y=[0.3, 0.5, 0.7, 1.10, 1.30, 1.50, 1.60, 1.75]



plt.plot(x, y)

plt.title('Crecimiento del ser humano')
plt.xlabel('Edad')
plt.ylabel('Altura')

plt.grid(color='green', ls='--', lw=2)
plt.show()
```

## MOSTRAR VARIAS GRAFICAS A LA VEZ

```python
#Vamos a usar el metodo subplot() para mostrar varias graficas a la vez
#en pantalla
#el metodo subplot() usa dentro de parentesis 3 argumentos, numero de filas
#numero de columnas y posicion en la pantalla del grafico (arriba a la
#izquierda) es el 1, arriba a la derecha es el 2
#plot 1:
x = [5, 10, 15, 20]
y = [-4, -8, -14, -30]

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = [5, 10, 15, 20]
y = [8, 16, 24, 32]

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
#vamos a cambiar los dos primeros parametros, en vez de 1, 2 ahora los
#hacemos 2, 1 y en vez de dos plots uno a lado de otro nos salen dos
#plots uno encima del otro
#plot 1:
x = [5, 10, 15, 20]
y = [-4, -8, -14, -30]

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = [5, 10, 15, 20]
y = [8, 16, 24, 32]

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

```python
#puedes aadir todos los subplots que quieras
x = [5, 10, 15, 20]
y = [4, 9, 12, 115]

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = [5, 10, 15, 20]
y = [15, 20, 25, 30]

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = [5, 10, 15, 20 ]
y = [3, 8, 1, 10]

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = [5, 10, 15, 20]
y = [10, 20, 30, 40]

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = [5, 10, 15, 20]
y = [3, 8, 1, 10]

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = [5, 10, 15, 10]
y = [10, 20, 30, 40]

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```

```python
#Vamos a ponerle titulo a cada subplot
#plot 1:
x = [3, 6, 9, 12]
y = [-4, -8, -14, -30]

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title('TEMPERATURA INVIERNO')

#plot 2:
x = [3, 6, 9, 12]
y = [8, 16, 24, 32]

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title('TEMPERATURA VERANO')

plt.show()
```

```python
#Se le puede aadir un titulo a todos los subplots usando el metodo
#suptitle()
#plot 1:
x = [3, 6, 9, 12]
y = [-4, -8, -14, -30]

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title('TEMPERATURA INVIERNO')

#plot 2:
x = [3, 6, 9, 12]
y = [8, 16, 24, 32]

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title('TEMPERATURA VERANO')

plt.suptitle('RUSIA')
plt.show()
```
